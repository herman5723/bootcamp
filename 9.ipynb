```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\tips.csv')
```

```python
smoking_counts = df['Smoking'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(smoking_counts, labels=smoking_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen'])
plt.title('Percentage of People in Smoking Area vs Non-Smoking Area')
plt.show()
```
