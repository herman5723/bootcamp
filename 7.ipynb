```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\ipl.csv')
```

```python
top_10_players = df.sort_values(by='total_runs', ascending=False).head(10)
```

```python
top_10_players = df.sort_values(by='total_runs', ascending=False).head(10)
#striker end
plt.figure(figsize=(10, 6))
plt.bar(top_10_players['batter'], top_10_players['total_runs'], color='skyblue')
plt.xlabel('Players')
plt.ylabel('Runs Scored')
plt.title('Top 10 Players by Runs Scored')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
top_10_players = df.sort_values(by='total_runs', ascending=False).head(10)
#non_striker end
plt.figure(figsize=(10, 6))
plt.bar(top_10_players['non_striker'], top_10_players['total_runs'], color='skyblue')
plt.xlabel('Players')
plt.ylabel('Runs Scored')
plt.title('Top 10 Players by Runs Scored')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
